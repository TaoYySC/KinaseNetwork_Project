# alf = 1 (raw)

```python
l2_map = {}
l2_map[1.000000013351432e-10] = '1e-10'
l2_map[1.0000000116860974e-07] = '1e-07'
l2_map[9.999999747378752e-05] = '1e-04'
l2_map[0.10000000149011612] = '1e-01'

perf_alf1 = pd.read_csv('result_LA_no_threshold/dp1_0_dp2_0/performance.csv', index_col=0)
perf_alf1['weight_decay'] = perf_alf1['weight_decay'].astype(np.float32).map(l2_map)
perf_alf1 = perf_alf1[(perf_alf1['weight_decay'] == '1e-10') & (perf_alf1['network_type']=='cpd')]
perf_alf1.index = range(perf_alf1.shape[0])

report_alf1 = perf_alf1[['weight_decay', 'drop_p_prior', 'drop_p_kki', 'cv']].drop_duplicates().copy()
report_alf1.index = range(report_alf1.shape[0])
for index, row in report_alf1.iterrows():
    tmp = perf_alf1.loc[(perf_alf1['weight_decay']==row['weight_decay'])&\
                      (perf_alf1['drop_p_prior']==row['drop_p_prior'])&\
                      (perf_alf1['drop_p_kki']==row['drop_p_kki'])&\
                      (perf_alf1['cv']==row['cv'])]
    # 在每个超参数组合下，每次交叉验证中，选取fit/refit（11个）模型R2最高时的AUPRC和R2作为这个超参数组合在这次交叉验证中的性能
    max_r2_row = tmp.loc[tmp['val_r2'].idxmax()]
    report_alf1.at[index, 'AUPRC'] = max_r2_row['AUPRC']
    report_alf1.at[index, 'R2'] = max_r2_row['val_r2']

    model = torch.load(f"result_LA_no_threshold/dp1_0_dp2_0/model/cv{max_r2_row['cv']}_wd9.999999999999999e-11_refit{int(max_r2_row['refit'])}.pth")
    model.eval()
    out = model(torch.tensor(np.array(data)).to(model.prior.fc.weight.dtype)).detach().numpy()
    out_flatten = out.flatten()
    r, _ = stats.pearsonr(raw_data_flatten, out_flatten)
    r_pct10, _ = stats.pearsonr(raw_data_flatten[(raw_data_flatten <= np.percentile(raw_data_flatten, 10)) & (raw_data_flatten >= 0)], 
                            out_flatten[(raw_data_flatten <= np.percentile(raw_data_flatten, 10)) & (raw_data_flatten >= 0)])
    r_pct25, _ = stats.pearsonr(raw_data_flatten[(raw_data_flatten <= np.percentile(raw_data_flatten, 25)) & (raw_data_flatten >= 0)], 
                            out_flatten[(raw_data_flatten <= np.percentile(raw_data_flatten, 25)) & (raw_data_flatten >= 0)])
    r_pct50, _ = stats.pearsonr(raw_data_flatten[(raw_data_flatten <= np.percentile(raw_data_flatten, 50)) & (raw_data_flatten >= 0)], 
                            out_flatten[(raw_data_flatten <= np.percentile(raw_data_flatten, 50)) & (raw_data_flatten >= 0)])
    r_pct75, _ = stats.pearsonr(raw_data_flatten[(raw_data_flatten <= np.percentile(raw_data_flatten, 75)) & (raw_data_flatten >= 0)], 
                                out_flatten[(raw_data_flatten <= np.percentile(raw_data_flatten, 75)) & (raw_data_flatten >= 0)])
    r_pct90, _ = stats.pearsonr(raw_data_flatten[(raw_data_flatten <= np.percentile(raw_data_flatten, 90)) & (raw_data_flatten >= 0)], 
                                out_flatten[(raw_data_flatten <= np.percentile(raw_data_flatten, 90)) & (raw_data_flatten >= 0)])

    report_alf1.at[index, 'Pearson'] = r
    report_alf1.at[index, 'Pearson_pct10'] = r_pct10
    report_alf1.at[index, 'Pearson_pct25'] = r_pct25
    report_alf1.at[index, 'Pearson_pct50'] = r_pct50
    report_alf1.at[index, 'Pearson_pct75'] = r_pct75
    report_alf1.at[index, 'Pearson_pct90'] = r_pct90

# 计算所有交叉验证的平均值代表每个超参数组合的性能
report_alf1 = report_alf1.groupby(['weight_decay', 'drop_p_prior', 'drop_p_kki'])[['AUPRC', 'R2', 'Pearson', 'Pearson_pct10', 'Pearson_pct25', 'Pearson_pct50', 'Pearson_pct75', 'Pearson_pct90']].mean().reset_index()

report_alf1
```

```python
best_cv = perf_alf1.sort_values(by='val_r2', ascending=False).iloc[0]['cv']
best_refit = perf_alf1.sort_values(by='val_r2', ascending=False).iloc[0]['refit']

print(best_cv, int(best_refit))

model = torch.load(f'result_LA_no_threshold/dp1_0_dp2_0/model/cv{best_cv}_wd9.999999999999999e-11_refit{int(best_refit)}.pth')
model.eval()
model_dict['alf1'] = model
```

```python
out = model_dict['alf1'](torch.tensor(np.array(data)).to(torch.float32)).detach().numpy()
out_flatten = out.flatten()

r, _ = stats.pearsonr(raw_data_flatten, out_flatten)
r_pct10, _ = stats.pearsonr(raw_data_flatten[(raw_data_flatten <= np.percentile(raw_data_flatten, 10)) & (raw_data_flatten >= 0)], 
                            out_flatten[(raw_data_flatten <= np.percentile(raw_data_flatten, 10)) & (raw_data_flatten >= 0)])
r_pct25, _ = stats.pearsonr(raw_data_flatten[(raw_data_flatten <= np.percentile(raw_data_flatten, 25)) & (raw_data_flatten >= 0)], 
                            out_flatten[(raw_data_flatten <= np.percentile(raw_data_flatten, 25)) & (raw_data_flatten >= 0)])
r_pct50, _ = stats.pearsonr(raw_data_flatten[(raw_data_flatten <= np.percentile(raw_data_flatten, 50)) & (raw_data_flatten >= 0)], 
                            out_flatten[(raw_data_flatten <= np.percentile(raw_data_flatten, 50)) & (raw_data_flatten >= 0)])
r_pct75, _ = stats.pearsonr(raw_data_flatten[(raw_data_flatten <= np.percentile(raw_data_flatten, 75)) & (raw_data_flatten >= 0)], 
                            out_flatten[(raw_data_flatten <= np.percentile(raw_data_flatten, 75)) & (raw_data_flatten >= 0)])
r_pct90, _ = stats.pearsonr(raw_data_flatten[(raw_data_flatten <= np.percentile(raw_data_flatten, 90)) & (raw_data_flatten >= 0)], 
                            out_flatten[(raw_data_flatten <= np.percentile(raw_data_flatten, 90)) & (raw_data_flatten >= 0)])

plt.figure(figsize=(10,5))

plt.subplot(121)
sns.regplot(x=raw_data_flatten, y=out_flatten, ci=None, line_kws={'color': 'red', 'lw': 1.5}, scatter_kws={'s': 1, 'alpha': 1})
plt.text(0.05, 0.95, f'Pearson r = {r:.2f}', transform=plt.gca().transAxes, verticalalignment='top')

plt.xlim(-0.1, 4.1)
plt.ylim(-0.1, 4.1)
plt.xlabel('Actual')
plt.ylabel('Reconstructed')
plt.title('alpha = 1 (raw)')


plt.subplot(122)
plt.bar(x=['pct10', 'pct25', 'pct50', 'pct75', 'pct90'],
        height=[r_pct10, r_pct25, r_pct50, r_pct75, r_pct90])
plt.ylabel('Pearson r')

for i, v in enumerate([r_pct10, r_pct25, r_pct50, r_pct75, r_pct90]):
    plt.text(i, v + 0.01, f'{v:.2f}', zorder=5, fontsize=9, ha='center')

plt.tight_layout()
```
